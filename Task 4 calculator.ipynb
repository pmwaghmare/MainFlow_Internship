```python
from tkinter import *
from PIL  import ImageTk,Image

def init_cal():
    t = Tk()
    t.title("Your Calculator")
    t.iconbitmap("D:\\Mainflow_Internship\\Task 4\\Calculator Photo\\Calc-icon.ico")
    t.geometry("300x430")

    bg = ImageTk.PhotoImage(file="D:\\Mainflow_Internship\\Task 4\\Calculator Photo\\300x430_background.png")
    bg_label = Label(t, image=bg)
    bg_label.place(x=0, y=0)

    e = Entry(t, width=30, borderwidth=8, font=("Times", 10))
    e.grid(row=0, column=0, pady=20, padx=30, columnspan=3)

    #load images
    images={
     "1":ImageTk.PhotoImage(file="D:\\Mainflow_Internship\\Task 4\\Calculator Photo\\1.png"),
     "2":ImageTk.PhotoImage(file="D:\\Mainflow_Internship\\Task 4\\Calculator Photo\\2.png"),
     "3":ImageTk.PhotoImage(file="D:\\Mainflow_Internship\\Task 4\\Calculator Photo\\3.png"),
     "4":ImageTk.PhotoImage(file="D:\\Mainflow_Internship\\Task 4\\Calculator Photo\\4.png"),
     "5":ImageTk.PhotoImage(file="D:\\Mainflow_Internship\\Task 4\\Calculator Photo\\5.png"),
     "6":ImageTk.PhotoImage(file="D:\\Mainflow_Internship\\Task 4\\Calculator Photo\\6.png"),
     "7":ImageTk.PhotoImage(file="D:\\Mainflow_Internship\\Task 4\\Calculator Photo\\7.png"),
     "8":ImageTk.PhotoImage(file="D:\\Mainflow_Internship\\Task 4\\Calculator Photo\\8.png"),
     "9":ImageTk.PhotoImage(file="D:\\Mainflow_Internship\\Task 4\\Calculator Photo\\9.png"),
     "0":ImageTk.PhotoImage(file="D:\\Mainflow_Internship\\Task 4\\Calculator Photo\\10.png"),
     "+":ImageTk.PhotoImage(file="D:\\Mainflow_Internship\\Task 4\\Calculator Photo\\11.png"),
     "-":ImageTk.PhotoImage(file="D:\\Mainflow_Internship\\Task 4\\Calculator Photo\\12.png"),
     "*":ImageTk.PhotoImage(file="D:\\Mainflow_Internship\\Task 4\\Calculator Photo\\13.png"),
     "/":ImageTk.PhotoImage(file="D:\\Mainflow_Internship\\Task 4\\Calculator Photo\\14.png"),
     "=":ImageTk.PhotoImage(file="D:\\Mainflow_Internship\\Task 4\\Calculator Photo\\15.png"),
     "C":ImageTk.PhotoImage(file="D:\\Mainflow_Internship\\Task 4\\Calculator Photo\\16.png"),
        
    }

    def btn_click(number):
        current = e.get()
        e.delete(0, END)
        e.insert(0, str(current) + str(number))

    def btn_add():
        num1 = e.get()
        global n1
        global math
        math = "addition"
        n1 = int(num1)
        e.delete(0, END)

    def btn_sub():
        num1 = e.get()
        global n1
        global math
        math = "substraction"
        n1 = int(num1)
        e.delete(0, END)

    def btn_mul():
        num1 = e.get()
        global n1
        global math
        math = "multiplication"
        n1 = int(num1)
        e.delete(0, END)

    def btn_div():
        num1 = e.get()
        global n1
        global math
        math = "division"
        n1 = int(num1)
        e.delete(0, END)

    def btn_equals():
        num2 = e.get()
        e.delete(0, END)
        if math == "addition":
            e.insert(0, n1 + int(num2))
        elif math == "substraction":
            e.insert(0, n1 - int(num2))
        elif math == "multiplication":
            e.insert(0, n1 * int(num2))
        elif math == "division":
            if int(num2) != 0:
                e.insert(0, n1 / int(num2))
            else:
                e.insert(0, "Error")

    def btn_clear():
        e.delete(0, END)

   #Defing buttons
    buttons = {
            "1": (1, 0, lambda: btn_click("1")),
            "2": (1, 1, lambda: btn_click("2")),
            "3": (1, 2, lambda: btn_click("3")),
            "4": (2, 0, lambda: btn_click("4")),
            "5": (2, 1, lambda: btn_click("5")),
            "6": (2, 2, lambda: btn_click("6")),
            "7": (3, 0, lambda: btn_click("7")),
            "8": (3, 1, lambda: btn_click("8")),
            "9": (3, 2, lambda: btn_click("9")),
            "0": (4, 1, lambda: btn_click("0")),
            "+": (4, 2, btn_add),
            "-": (4, 0, btn_sub),
            "*": (5, 1, btn_mul),
            "/": (5, 0, btn_div),
            "=": (5, 2, btn_equals),
            "C": (6, 1, btn_clear)
           }

    for btn, (row,col,cmd) in buttons.items():
       Button(t,border="3",image=images[btn],command=cmd).grid(row=row,column=col)

    #keep references to images
    t.images=images

    t.mainloop()

#Initialize the calculator
init_cal()
      
```
