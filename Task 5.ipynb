```python
import tkinter as tk
from tkinter import *
from tkinter import ttk
from datetime import datetime
import requests
from PIL import ImageTk, Image
from tkinter import messagebox

root=tk.Tk()
root.geometry("600x270")
root.title("currency converter")
root.iconbitmap("D:\\Mainflow_Internship\\Task 5\\icon.ico")
root.maxsize(600,270)
root.minsize(600,270)

image=Image.open("D:\\Mainflow_Internship\\Task 5\\currency.png")
zoom=0.5

#multiple image size
pixels_x,pixels_y = tuple([int(zoom * x) for x in image.size])

img=ImageTk.PhotoImage(image.resize((pixels_x,pixels_y)))

# Create a PhotoImage object directly
photo = PhotoImage(file="D:\\Mainflow_Internship\\Task 5\\currency.png")

panel = Label(root,image = photo)
panel.image = photo
panel.place(x=150, y=65)

t1 = Entry(root, width=20, borderwidth=1, font=('roboto', 10, 'bold'))
t1.place(x=20, y=60)

def show_data():
           global t1
           amount=t1.get()
           from_currency =c1.get()
           to_currency=c2.get()
           url = "https://api.currencylayer.com/live?access_key=4273d2c37f738367f08780b934ce7dda&format=1"

           if amount =="" :
                messagebox.showerror("Currency Converter", "Please fill the amount")
           elif to_currency == "" :
                 messagebox.showerror("Currency Converter", "Please choose the currency")

           else:
                data =requests.get(url).json()
                currency = from_currency.strip()+to_currency.strip()
                amount = int(amount)
                cc =data['quotes'][currency]
                cur_conv == cc*amount
                t2.insert(0,cur_conv)

                text.insert("end", f'{amount} united state dollar equals {cur_conv} {to_currency} \n\m Last time update .... \t{datetime.now()}')

def clear():
    E1.delete(0,'end')
    E2.delete(0,'end')
    text.delete(1.0,'end')

l1 =Label(root , text ="USD Currency Converter using Python", font =('vardana','10','bold'))
l1.place(x=150,y=15)

amt=Label(root,text="amount",font=('roboto',10,'bold'))
amt.place(x=20,y=15)

E1=Entry(root,width=20,borderwidth=1,font=('roboto',10,'bold'))
E1.place(x=20,y=60)

c1=tk.StringVar()
c2=tk.StringVar()

currencychoose1=ttk.Combobox(root,width=20, textvariable = c1,state='readonly',font=('vardana',10,'bold'))

#adding combobox drop down list
currencychoose1['values']=(
                            'USD',
)
currencychoose1.place(x=300,y=40)
currencychoose1.current(0)


E2=Entry(root,width=20,borderwidth=1,font=('roboto',10,'bold'))
E2.place(x=20,y=80)

currencychoose2=ttk.Combobox(root,width=20, textvariable =c2,state='readonly',font=('vardana',10,'bold'))

#adding combobox drop down list
currencychoose2['values']=(
                            'ALL',
                            'ARS',
                            'INR',
                            'AUD',
                            'BSD',
                            'EUR',
                            'GBP', 
                            'JPY',
                            'CHF',
                            'CAD', 
                            'AUD',
                            'YER',
)
currencychoose2.place(x=300,y=140)
currencychoose2.current(0)

text = Text(root,height=7, width=52, font=('vardana','10','bold'))
text.place(x=100,y=180)

b=Button(root,text="Search",command=show_data, font=('vardana',10,'bold'),borderwidth=2,bg="red",fg="white")
b.place(x=20,y=120)

clear=Button(root,text="clear",command=clear, font=('vardana',10,'bold'),borderwidth=2,bg="blue",fg="white")
clear.place(x=20,y=170)

root.mainloop()

```
